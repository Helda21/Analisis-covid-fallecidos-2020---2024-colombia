```python
conteoyear1 = df['AñoInicioSintomas'].value_counts()
```

```python

# Identifica los años que tienen un número muy pequeño de personas
umbral = 1000  # Define el umbral para considerar que un año tiene un número muy pequeño de personas
anos_pequenos = [ano for ano, personas in conteoyear1.items() if personas < umbral]

# Suma el número de personas para los años pequeños
personas_anos_pequenos = sum([personas for ano, personas in conteoyear1.items() if ano in anos_pequenos])

# Crea un nuevo diccionario excluyendo los años pequeños y agrega la categoría "Otros años posteriores"
conteo_modificado = {ano: personas for ano, personas in conteoyear1.items() if ano not in anos_pequenos}
conteo_modificado['Otros años posteriores'] = personas_anos_pequenos

# Grafica el nuevo diccionario con el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(conteo_modificado.values(), labels=conteo_modificado.keys(), autopct='%1.1f%%', startangle=60)
plt.xlabel('Año')
plt.ylabel('Número de personas con síntomas')
plt.title('Cantidad de personas por año con inicio de síntomas')
plt.show()
```

```python
# Calcular la cantidad de casos por departamento
casos_por_departamento = df['Nombre departamento'].value_counts()

# Graficar un gráfico de barras
plt.figure(figsize=(10, 6))
casos_por_departamento.plot(kind='bar')
plt.xlabel('Departamento')
plt.ylabel('Cantidad de casos')
plt.title('Distribución de casos por departamento')
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x para mayor legibilidad
plt.tight_layout()  # Ajustar el diseño para evitar cortar las etiquetas
plt.show()
```

```python
# Definir los límites de los grupos de edad y las etiquetas
limites_edad = [0, 18, 35, 50, float('inf')]  # Rango de edad de 0 a 18, de 19 a 35, de 36 a 50 y más de 50
etiquetas_edad = ['Adolescente', 'Joven', 'Adulto', 'Adulto Mayor']

# Categorizar las edades en los grupos definidos
categorias_edad = pd.cut(df['Edad'], bins=limites_edad, labels=etiquetas_edad)

# Contar los casos en cada grupo de edad
casos_por_grupo_edad = categorias_edad.value_counts()

# Graficar un gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(casos_por_grupo_edad, labels=casos_por_grupo_edad.index, autopct='%1.1f%%', startangle=60)
plt.xlabel('Grupo de Edad')
plt.ylabel('Cantidad de casos')
plt.title('Distribución de casos por grupo de edad')
plt.show()
```
